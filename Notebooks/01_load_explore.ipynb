```python
RANDOM_SEED = 42

import random
import numpy as np
import pandas as pd

random.seed(RANDOM_SEED)          # Python built-in
np.random.seed(RANDOM_SEED)       # NumPy
pd.set_option("future.no_silent_downcasting", True)


from pathlib import Path

# Your notebook is in: Project/Notebooks/
# So go up one level to reach Project/ (the real root)
PROJECT_ROOT = Path.cwd().parent

RAW_ROOT      = PROJECT_ROOT / "Raw Project Datasets"
CAC_DIR       = RAW_ROOT / "CAC-LTV Model Analysis for SaaS Business Insights"
STARTUPS_DIR  = RAW_ROOT / "Business Startups Data on SAAS products"
DATA_DIR      = PROJECT_ROOT / "data" / "raw"

print("Using project root:", PROJECT_ROOT)
print("CAC file path:", CAC_DIR / "cac_ltv_model.csv")


def load_csv(path: Path, **kwargs) -> pd.DataFrame:
    if not path.is_file():
        raise FileNotFoundError(f"\nFile NOT FOUND:\n    {path.resolve()}\n")
    return pd.read_csv(path, low_memory=False, **kwargs)


# Load every CSV
df_cac          = load_csv(CAC_DIR      / "cac_ltv_model.csv")
df_startups     = load_csv(STARTUPS_DIR / "saas-businesses-data.csv")

df_subscriptions = load_csv(DATA_DIR / "subscriptions.csv")
df_usage         = load_csv(DATA_DIR / "usage.csv")
df_support       = load_csv(DATA_DIR / "support_tickets.csv")
df_revenue       = load_csv(DATA_DIR / "revenue.csv")
df_churn         = load_csv(DATA_DIR / "churn.csv")

df_sub = (
    df_subscriptions
    .merge(df_usage,    on="user_id", how="left")
    .merge(df_support,  on="user_id", how="left")
    .merge(df_revenue,  on="user_id", how="left")
    .merge(df_churn,    on="user_id", how="left")
)
```
